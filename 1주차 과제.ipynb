_네이버의 파파고_

파파고는 네이버의 실시간 번역 앱이다. 인공신경망 번역(NMT)이 적용 됐다.
NMT는 Neural Machine Translation(인공신경망 기반 기계번역)의 약어로, 인공신경망
이라고 하는 커다란 가상공간을 통해 문장을 이해하고 번역한다.파파고의 NMT 기술은 
입력 문장을 문장벡터로 변환하는 신경망(encoder)과 문장벡터에서 번역하는 언어의 
문장을 생성하는 신경망(decoder)을 대규모의 병렬 코퍼스부터 자동으로 학습한다.
입력문장의 일부가 아니라 문장 전체 정보를 바탕으로 번역을 수행하기 때문에 문장의
맥락을 이해하고, 보다 자연스러운 결과를 만들게 된다. 
높임말 번역 기능은 고유의 높임말 체계를 갖춘 한국어 특성을 Papago AI학습 
알고리즘에 반영하여, 번역시 문장에 높임말을 적용할 수 있는 기능을 제공한다. 
Papago NMT는 네이버 서비스를 통해 쌓인 대규모의 데이터를 매일 자동으로 학습한다.
이렇게 생성된 번역 모델을 통해,기존 SMT방식의 번역보다 더욱 정확하고 문장 맥락에
맞는 번역을 하는 것이 특징이다.

* __음성__

_구글의 타코트론_

타코트론은 구글의 음성 합성의 대표적인 모델이다. 딥러닝으로 입력과 출력 사이의 
비선형적인 관계를 수많은 뉴런 모델들이 자동으로 발견하고, 학습한다. 
딥러닝 기술 기반에서 입력과 출력 데이터만 있으면 기계가 자동으로 중간 과정을 
찾아서 모델링하고 모델값을 저장해 음성을 바로 합성한다.
타코트론은 CBHG라는 신경망 블록을 사용하는데 convolutional 1-D filters, bank, 
highway networks, gated recurrent unit bidirectional의 첫 글자를 따서 지어진 
이름으로 단어가 나열된 순서대로 각 블록을 쌓아 만든 구조이기 때문이다. 부분적인 
특징을 잡아내는데 유용한 CNN(convolutional neural network)과 연속적인(sequential)
데이터를 생성하기 좋은 RNN(recurrent neural network)의 특징을 모두 활용했다. 
수초짜리의 발화 데이터를 정해진 크기만큼 순차적으로 학습하고, 목소리 특징을 뽑아
내는(CNN) 동시에, 뽑혀진 부분 특징을 전체적으로 보면서 결과를 만들어 낸다(RNN).
또한 문자열을 입력으로, 음성의 스펙트럼 특징벡터 열을 출력으로 지정해 문자열이 
스펙트럼으로 변환되는 중간 과정을 자동으로 학습한다. 이때 입력과 출력의 길이의 
차이가 발생하는데, 어텐션(attention) 이라는 신경망 층을 도입해 입력과 출력 사이의
매핑(Mapping) 관계를 학습한다.encoder란 텍스트를 입력 받았을 때 이를 잘 나타내는
숫자로 변환시켜 주는 것이다. 숫자로 변환된 텍스트는 decoder 과정을 거친다. 그리고 
attention에서 encoder의 결과와 decoder를 서로 적절히 이어주는 역할을 수행한다. 
실질적으로 서로를 이어주는 역할이다. attention을 성공적으로 사용하면, decoder가 
encoder의 상태를 순차적으로 집중하게 된다. 그래서 끊임없이 가지런한 attention을 
그릴 수 있다. 딥러닝을 바탕으로 한 attention 모듈 덕분에, 기계는 스스로 여러 
데이터를 학습하며 화자의 억양이나 말투를 그대로 구사할 수 있다.

* 이미지

_페이스북의 얼굴 인식_

얼굴 인식은 Facebook상에 회원이 나온 것 같은 사진 및 동영상(예: 회원의 프로필 사진, 회원이 태그되어 있는 사진과 동영상)을 분석하여 ‘템플릿’이라고 하는 고유한 수치로 만드는 데 사용되는 기능이다. 얼굴 인식 설정을 켜면 Facebook은 회원의 템플릿을 생성하고 다른 사진, 동영상 및 카메라가 사용된 다른 부분(라이브 방송 등)과 비교하여 해당 콘텐츠에 회원이 나타나는지 인식한다.
페이스북 인공지능 연구소는 ‘딥 페이스’ 알고리즘을 개발했다. 사진 속 얼굴을 분석해 같은 사람을 연결해주며 정확도는 97.25%수준이며 인간의 97.53%와 비슷하다.
딥 페이스는 불규칙한 데이터를 이해하는 인공지능 프로그램 딥러닝을 사용해 정확도를 보다 높였다. 얼굴을 더 정확히 분석하기 위해 연구자들은 동물의 중추신경계를 흉내 낸 ‘신경망 분석’을 만들었다.
MIT의 테크놀러지리뷰는 딥페이스가 ‘가상으로 얼굴을 회전시킨 3D 모델’을 쓴다고 분석했다. 카메라에 비친 얼굴을 바탕으로 입체를 유추해내는 것이다. 평균적인 얼굴 전면부 3D 모델을 사용해 얼굴 각도를 수정해낸다. 유사한 이미지의 다른 사진에서 가상의 원형 얼굴을 비교해낼 수 있다. 1억2000만가지 연관관계를 파악해내는 9개의 계층에 대조하며 4000여명의 사용자 얼굴에서 추출한 400만개 얼굴 이미지 데이터를 활용한다.
대량의 데이터 속 패턴을 인지해 내는 학습 기능을 가진 일종의 ‘모의 신경망’ 네트워크가 핵심이다. 여기에 인공지능이 쓰인다.
페이스북 연구원들은 다른 얼굴을 구별해 내는 이 인공지능 기술을 얼굴인식이 아닌 ‘얼굴확인(facial verification)’ 과정이라 부른다. 확인을 거친 후 기존 얼굴인식 기술로 보강하는 것이다.그간 페이스북은 눈과 코 거리 등을 파악하는 인식기술을 썼다. 유사한 얼굴 구조를 가진 사람을 잘못 태깅하는 오류가 종종 일어났지만 이 부분을 딥페이스로 상당히 수정했다.


* 자율주행

_현대자동차_

현대자동차와 기아자동차는 운전자의 주행성향에 맞는 부분 자율주행을 구현하는 기술
인 SCC-ML을 세계 최초로 개발했다. 스마트 크루즈 컨트롤SCC은 앞차와의 거리를 일정
하게 유지해 운전자가 설정한 속도로 자유주행을 해주는 기능으로 ADAS(첨단 운전자보
조 시스템)의 주요 기술 중 하나다. SCC-ML은 여기에 AI기술을 더해 운전자의 주행성향
을 차가 스스로 학습해 SCC 작동 시 운전자와 거의 흡사한 패턴으로 자율주행을 해준다.
기존의 SCC는 앞차와의 거리, 가속성 등의 주행패턴을 운전자가 직접 설정해야 했으며
조절되는 단계가 세밀하지 않아 운전성향을 고스란히 반영할 수 없었다. 
이번 현대차·기아차가 독자 개발한 SCC-ML의 원리는 먼저 전방카메라, 레이더 등의 
센서가 다양한 운전상황에서 발생되는 정보를 지속적으로 수집해 ADAS의 두뇌격인 
제어컴퓨터로 보낸다. 제어컴퓨터는 입력된 정보로부터 운전자의 주행습관을 추출해 
종합적인 주행성향을 파악한다. 이때 인공지능 기술 중 하나인 머신러닝 알고리즘이 
적용된다. SCC-ML은 다양한 상황을 고려해 총 만개 이상의 패턴을 구분함으로써 
어떤 운전자의 성향에도 맞출 수 있는 SCC 기술을 개발했다.
